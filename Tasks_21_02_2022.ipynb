```python
###Inserting CSV into MONGODB###
def read_csv_and_write():
    logging.info("Reading csv file")
    try:
        with open("carbon_nanotubes.csv","r") as File:
            carbon_nanotubes=csv.reader(File,delimiter='\n')
            logging.info("Getting the header record for using as keys")
            carb=next(carbon_nanotubes)
            header=carb[0].split(";")
            print(header)        
            logging.info("Skipping the first line from csv")
            li =[]
            for i in carbon_nanotubes:
                li= i[0].split(";")
                dict1=dict(zip(header, li))
                print(dict1)
                Coll.insert_one(dict1)
        logging.info("File has been read successfully and uploaded to MongoDB")
    except Exception as e :
        logging.error("Error has happened")
        logging.exception("Exception occured"+ str(e))
```

```python
###Independent insertion of data###
def individual_data_insert():
    logging.info("Inserting individual record")
    try:
        dict5 = {'Chiral indice n': '13', 'Chiral indice m': '20', 'Initial atomic coordinate u': '0,679005', 'Initial atomic coordinate v': '0,701318', 'Initial atomic coordinate w': '0,017033', "Calculated atomic coordinates u'": '0,721039', "Calculated atomic coordinates v'": '0,730232', "Calculated atomic coordinates w'": '0,017014'}
        Data_insertion=Coll.insert_one(dict5)
        logging.info("Insertion Successfull")
    except Exception as e :
        logging.error("Error has happened")
        logging.exception("Exception occured"+ str(e))
```

```python
##Updation of Data by changing Chiral indice m 2 to 21 of all records with Chiral indice m as 2####
logging.info("Updating records")
def update_records():
    try:
        Updation_Data=Coll.update_many({"Chiral indice m":"2"},{"$set":{"Chiral indice m":"21"}})
        logging.info("Updation Successfull")
    except Exception as e :
        logging.error("Error has happened")
        logging.exception("Exception occured"+ str(e))
```

```python
###Deletion of all data with Chiral indice m as 21####
logging.info("Deleting records")
def delete_records():
    try:
        Deletion_Data=Coll.delete_many({"Chiral indice m":"21"})
        logging.info("Deletion Successfull")
    except Exception as e :
        logging.error("Error has happened")
        logging.exception("Exception occured"+ str(e))
```

```python
###find data with Chiral indice n equal to '10' and '11'####
logging.info("Finding records with Chiral indice n  in 10,11")
def find_records():
    try:
        Find_Data=Coll.find({"Chiral indice n":{'$in':["10","11"]}})
        logging.info("Find Successfull")
    except Exception as e :
        logging.error("Error has happened")
        logging.exception("Exception occured"+ str(e))    
    for i in Find_Data:
        print(i)
```

```python
###Filter data with Initial atomic coordinate v having 2 values####
logging.info("Filter records with Initial atomic coordinate v having 2 values")
def filter_records():
    try:
        search_filter={"Initial atomic coordinate v":{"$in":['0,668895','0,430299',]}}
        Find_Data=Coll.find(search_filter)
        logging.info("Filter Successfull")
    except Exception as e :
        logging.error("Error has happened")
        logging.exception("Exception occured"+ str(e))
    for i in Find_Data:
        print(i)
```
